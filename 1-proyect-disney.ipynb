```python
# Cleaning number columns
df['date_added'] = pd.to_datetime(df['date_added'], infer_datetime_format=True)
df['release_year'] = df['release_year'].astype(int)
df['show_id'] = df['show_id'].replace({'s':''}, regex=True)
```

```python
df.duration.unique()
```

```python
# Extract numeric values
df["duration-number"] = df["duration"].str.extract('(\d+)')
df["duration-type"] = df["duration"].str.extract('([a-zA-Z]+)')
df['duration-number'] = df['duration-number'].astype(int)
df
```
